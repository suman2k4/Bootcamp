```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import pandas as pd
df=pd.read_csv('C:/Users/HP/Downloads/archive/bollywood.csv')
df.head()
km=KMeans(n_clusters=5)
km
init_methods = ['random', 'k-means++']
for init in init_methods:
    kmeans = KMeans(n_clusters=3, init=init, random_state=42)
    kmeans.fit(df[['Revenue(INR)','Budget(INR)']])
    plt.scatter(df['Budget(INR)'], df['Budget(INR)'], c=kmeans.labels_, cmap='viridis', alpha=0.6)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200)
    plt.title(f"K-means with {init} initialization")
    plt.show()
```
